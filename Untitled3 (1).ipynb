## <a> **INTRODUCTION** </a>

This data set contains the information of a population sample from urban and rural locations with various lifestyle choices and classifying them as healthy or not. The population size of data set is **11514** and has total **8** variables.

### <a> **LOADING LIBRARIES** </a>

```python
import pandas as pd
#import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split, RandomizedSearchCV, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.tree import DecisionTreeClassifier
#from sklearn.impute import SimpleImputer
import lightgbm as lgb
from lightgbm import LGBMClassifier
import xgboost
from xgboost import XGBClassifier, XGBRFClassifier
from xgboost import plot_tree, plot_importance
from sklearn.model_selection import KFold
```

### <a>**LOAD DATA**

```python
train_data = pd.read_csv('/content/Train_Data (1).csv')
test_data = pd.read_csv('/content/Test_Data (1).csv')

# viewing data
train_data.head()
```

### <a>**DATA CLEANING AND DIMENSION SELECTION**

```python
# important dimensions + PCA
le = LabelEncoder()

# DATA CLEANING AND MINING
#train_data['Food preference'] = le.fit_transform(train_data['Food preference'])
train_data['Smoker?'] = le.fit_transform(train_data['Smoker?'])
train_data['Living in?'] = le.fit_transform(train_data['Living in?'])
#train_data['Any heriditary condition?'] = le.fit_transform(train_data['Any heriditary condition?'])
train_data = train_data.drop(['Food preference','Any heriditary condition?'], axis=1)
train_data=train_data.fillna(0)
#imputer = SimpleImputer(strategy='median')
#train_data = imputer.fit_transform(train_data)
train_data.head()
```

###<a> **DATA SPLIT - STANDARDIZATION - PCA**

```python
#split into input data x and output data y
predictors = train_data.drop (['Healthy'], axis=1)  #.iloc[:,0:15]
outcomes = train_data['Healthy']      #.iloc[:,15]

#standardizing
sd = StandardScaler()
stand_train_data = sd.fit(predictors)
predictors = stand_train_data.transform(predictors)

#using pca
pca = PCA(n_components=8)
predictors_1 = pca.fit(predictors)
predictors = predictors_1.transform(predictors)
predictors.shape
```

```python
#data split
x_train_data = predictors
y_train_data = outcomes

#train test split of train data
X_train,X_test,y_train,y_test = train_test_split(x_train_data,y_train_data, test_size=0.2)
```

## <a>**MODEL SELACTION AND TESTING**



### <a> **1. RANDOM FOREST CLASSIFIER**



```python
#fitting the model
# USING RANDOM FORREST CLASSIFIER
rf_class = RandomForestClassifier()
rf_class.fit(X_train,y_train)
y_pred = rf_class.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
print('Accuracy :', accuracy)
```

### <a>**DESCISION TREE CLASSIFIER**

```python
# predictions using descision tree classifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
y_pred2 = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred2)
print('Accuracy :', accuracy)
```

### <a>**EXTREME GRADIESNT BOOST**

```python
# using xgb for evaluation

#DEFINING A FUNCTION TO SMOOTHEN THE PROCESS
def score(model, title = "Default"):
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
#     print(confusion_matrix(y_test, preds))
    accuracy = round(accuracy_score(y_test, preds), 5)
    print('Accuracy for', title, ':', accuracy, '\n')



xgb_model = XGBClassifier(n_estimator=1000, learning_rate=0.01)
score(xgb_model, "extreme GB model")
```
